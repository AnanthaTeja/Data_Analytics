```python
county_df=sale_df2.groupby("COUNTRY").sum()
county_df
```

Cleaning and Adding


```python
county_df.drop(['ORDERNUMBER','YEAR_ID',"MONTH_ID",],axis=1,inplace=True)
county_df['Total'] =  county_df['QUANTITYORDERED']+county_df['SALES']+county_df["MSRP"]
county_df
```

```python
county_df.sort_values(by='Total',ascending=False)
```

```python
sns.regplot(x='ORDERLINENUMBER', y='QTR_ID', data=county_df,color='orange',marker='+')
```
